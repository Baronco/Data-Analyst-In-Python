## Categories for institute_service Column

first transformed the different institute_service patterns:

```python
combined['institute_service'] = combined['institute_service'].astype(str).copy()
combined['institute_service'].value_counts(dropna=False).head()
```

I changed the category 'Less than 1 year' and 'More than 20 years' for its numeric value:

```python
combined.loc[combined['institute_service'] == 'Less than 1 year','institute_service'] = 1
combined.loc[combined['institute_service'] == 'More than 20 years','institute_service'] = 20
combined['institute_service'].value_counts(dropna=False).head()
```

Here I extracted the patterns from the institute_service:

```python
service_pattern = r"(?P<inf_year>[0-9]+)(-)(?P<sup_year>[0-9]+)|(?P<year>[0-9]+)"
service_years = combined['institute_service'].str.extract(service_pattern,expand=True)
service_years.head(12)
```

```python
service_years.tail(12)
```

Those whose institute_service time equals a range was averaged and rounded to an integer value:

```python
bool_years = service_years['inf_year'].notnull() & service_years['sup_year'].notnull()
service_years.loc[bool_years,'year'] = round(0.5*(service_years['inf_year'].astype(float)+service_years['sup_year'].astype(float)),0)
service_years.head(12)
```

To finish I modified the values to be able to create the categories:
- New: Less than 3 years at a company
- Experienced: 3-6 years at a company
- Established: 7-10 years at a company
- Veteran: 11 or more years at a company

```python
combined['institute_service'] = service_years['year'].astype(float).copy()
combined['institute_service'].value_counts(dropna=False)
```

Now the categories are created from the instute_service time:

```python
def categories(col):
    if pd.isnull(col) is True:
        pass
    elif col < 3:
        return 'New'
    elif 3 <= col <= 6:
        return 'Experienced'
    elif (7 <= col) and (col <= 10):
        return 'Established'
    else:
        return 'Veteran'
    
combined['service_cat'] = combined['institute_service'].apply(categories)
```

# Analysis of results

- I used the DataFrame.fillna() method to replace the missing values in the dissatisfied column with the value that occurs most frequently in this column, in this case 'False'.

- I used the DataFrame.fillna() method to replace the missing values in the service_cat column with the value that occurs most frequently in this column, in this case 'Experienced '.


```python
combined['dissatisfied'] = combined['dissatisfied'].fillna(False)
combined['dissatisfied'].value_counts(dropna=False)
```

```python
combined['service_cat'] = combined['service_cat'].fillna('Experienced')
combined['service_cat'].value_counts(dropna=False)
```
